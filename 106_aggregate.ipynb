# C* with N*

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from collections import Counter
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = 'all'
```

```python
df_txn_full = pd.read_csv('./data/002_txn.csv')
df_txn_full.head()
```

```python
c, n = 'c5', 'n3' # select 
```

```python
df_txn = df_txn_full[['id', c, n]].copy()
df_txn.head()
```

```python
cn_nn = df_txn.groupby([c])[[n]].mean().reset_index()
cn_nn.columns = [c, f'{c}_{n}']
cn_nn.head()
```

```python
df_txn = pd.merge(df_txn, cn_nn, on=c)
```

```python
df_txn.head()
```

```python
df_txn.drop([c, n], axis=1, inplace=True)
```

```python
tmp = df_txn.groupby(['id']).mean()
```

```python
tmp.head()
```

```python
tmp.to_csv(f'./data/100_{c}_{n}_mean_mean.csv')
```
