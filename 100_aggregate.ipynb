# One-hot c5

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from collections import Counter
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = 'all'
```

```python
df_txn = pd.read_csv('./data/002_txn.csv')
df_txn.head()
```

```python
df_txn = df_txn[['id', 'c5']]
```

```python
df_txn.head()
```

```python
dummies = pd.get_dummies(df_txn['c5'], prefix='c5_')
```

```python
dummies.head()
```

```python
df_txn = pd.concat(
    [
        df_txn, dummies
    ], axis=1
)
```

```python
df_txn.head()
```

```python
df_txn.drop('c5', axis=1, inplace=True)
```

```python
tmp = df_txn.groupby(['id']).sum()
```

```python
tmp.head()
```

```python
tmp.to_csv('./data/100_c5_encode.csv')
```
