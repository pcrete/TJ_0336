```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from collections import Counter
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = 'all'
```

```python
df_dm = pd.read_csv('./data/001_demo.csv')
df_dm.head()
```

```python
df_dm.isna().sum()*100/len(df_dm)
```

```python
# fill `n2` & `c1` with median
```

```python
df_dm['n2'] = df_dm['n2'].fillna(df_dm['n2'].median())
```

```python
df_dm.head()
```

```python
df_dm['c1'] = df_dm['c1'].fillna(df_dm['c1'].median())
```

```python
df_dm.head()
```

```python
df_dm.isna().sum()*100/len(df_dm)
```

```python
df_dm.drop('n1', axis=1, inplace=True)
```

```python
df_dm.head()
```

```python
df = pd.concat(
    [
        df_dm.drop(['c0', 'c1', 'c2', 'c3', 'c4'], axis=1),
        pd.get_dummies(df_dm['c0'].astype(int), prefix='c0'),
        pd.get_dummies(df_dm['c1'].astype(int), prefix='c1'),
        pd.get_dummies(df_dm['c2'].astype(int), prefix='c2'),
        pd.get_dummies(df_dm['c3'].astype(int), prefix='c3'),
        pd.get_dummies(df_dm['c4'].astype(int), prefix='c4')
        
    ], axis=1
)

df.head()
```

```python
df.to_csv('./data/100_demo.csv', index=False)
```
