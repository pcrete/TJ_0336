```python
df_dm    = pd.read_csv('./data/001_demo.csv')
df_txn   = pd.read_csv('./data/002_txn.csv')
df_train = pd.read_csv('./data/003_train.csv')
df_test  = pd.read_csv('./data/004_test.csv')
```

```python
df_txn.head()
```

```python
columns = ['id','n4','n5','n6','n7']
df_txn_n = df_txn[columns].copy()
df_txn_n.head()
```

```python
df_txn_ns = pd.concat([
    df_txn_n.groupby('id').mean(), df_txn_n.groupby('id').max(),
    df_txn_n.groupby('id').min(),  df_txn_n.groupby('id').median(),
    df_txn_n.groupby('id').std()
], axis=1)

cols = df_txn_n.columns[1:].values
df_txn_ns.columns = list(cols+'_mean') + list(cols+'_max')+\
                    list(cols+'_min')  + list(cols+'_median')+list(cols+'_std')

df_txn_ns.head()
df_txn_ns.shape
```

```python
df_txn_ns.to_csv('./data/101_agg_txn_ns.csv')
```

```python
columns = ['id', 'old_cc_label', 'c5', 'c6', 'c7']

df_txn_c = df_txn[columns].copy()
df_txn_c.head(100)
```

```python
tmp = df_txn_c[['id','c5', 'c6', 'c7']].groupby('id').agg(lambda x:x.value_counts().index[0])
tmp.head()
```

```python
df_txn_cmj = tmp.copy()
df_txn_cmj.columns = df_txn_cmj.columns.values+'_mj'
df_txn_cmj.head()
```

```python
columns = ['id', 'old_cc_label']

df_txn_clb = df_txn[columns].copy()
df_txn_clb_oh = pd.get_dummies(df_txn_clb['old_cc_label'], prefix='old_cc_label')

df_txn_clb_oh = pd.concat([
    df_txn_clb,
    df_txn_clb_oh
], axis=1)

df_txn_clb_oh = df_txn_clb_oh.drop('old_cc_label', axis=1)
df_txn_clb_oh = df_txn_clb_oh.groupby(['id']).mean()
df_txn_clb_oh.head()
```

```python
df_txn_clb_oh.columns
```

```python
df_txn_all = pd.concat([
    df_txn_ns,
    df_txn_cmj,
    df_txn_clb_oh
], axis=1)
```
